## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_nodes = len(X_df_processed.columns)

# Create the input layer
input_layer = layers.Input(shape=(len(X_df_processed.columns)), name='input_layer')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(64, activation='relu', name='department_hidden')(shared_layer2)

# Create the output layer
department_output = layers.Dense(3, activation='softmax', name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(64, activation='relu', name='attrition_hidden')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(1, activation='sigmoid', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_df_processed,
    {'department_output': y_df_department, 'attrition_output': y_df_attrition},
    epochs=100,
    batch_size=32,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_df_processed_test, {'department_output': y_df_department_test, 'attrition_output': y_df_attrition_test})
test_results
```
